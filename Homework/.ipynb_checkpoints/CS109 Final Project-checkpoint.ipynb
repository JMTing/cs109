## Motivation
Swimming is a sport in which people race one another in the pool and improve one's time in a given event. In competitive swimming, swimmers would swim a lot of these events throughout the season. However, swimmers typically do high intensity training in the beginning and middle of the season to build up their strength and endurance, so the times that they go usually is not their personal best as their body is not in peak physical condition. At the end of the season, the workload in practice decreases significantly in a phase called tapering, and the swimmers swims their races at their optimal level at their respective championship meet. Just like with any competitive race, there is a basis on which swimmers are ranked by their personal best times, but usually swimmers go faster than their personal best at the championship meet, so their new personal best time and the outcome of the race is unknown until it happens. 

We hope to collect this data and interpret it to gain a quantitative insight on predicting the times of what the men's Ivy League swimmers can go at the end of the season, which we will then use to simulate the Men's Ivy League Swimming Championship results.
