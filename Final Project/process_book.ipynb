## Overview and Motivation

Having gone through the collegiate swim recruiting process, the three of us noticed that most of the recruiting was focused largely on subjective evidence and reasoning from most collegiate coaches. What was most striking to us was that the top 30 NCAA Division 1 swimming schools were recruiting the same 100-150 kids each year. We realized that current ranking metrics and models for recruitment do not take into account the significant amount of data that is freely available, and we began to wonder whether we could use this data to help build better swim outcome prediction models. Whether the swimming outcome is the future progression of a recruit, or the projected times of a current collegiate swimmer at any given swim meet.

With a sufficiently sophisticated model, collegiate coaches could use projected time progressions and swimmer developments to predict the outcomes of upcoming meets, adjust meet entries and strategies accordingly, and plan for future contingencies (i.e. recruiting a strong swimmer in a certain event after the graduation of a current swimmer). Additionally, the model could potentially be used to tailor recruitment suggestions to different schools. This way, a coach could focus on recruiting otherwise overlooked swimmers who have significant potentially for success, but who are not currently fast enough to be highly sought after. This would enable recruiting to delve deeper into the ranks of prospective collegiate swimmers, and also hopefully allow school to recruit better classes by focusing on recruits who would excel in their own program.

## Initial Questions

Swimming is a uniquely quantitative sport in that all races are conducted under very similar and controlled conditions, and the data is recorded and freely available (although not easily accessible) from the national governing body, USA Swimming. USA Swimming maintains a searchable times database for every swim that is registered within the country. This includes all high school, club, and collegiate competitions.

We wanted to know whether any significant or relevant trends could be discovered from aggregating a significant body of data and performing analysis that we (as swimmers) perceived to be relevant.

The questions we hope to answer with the study of this data are as follows:

1. Can a swimmers future times be predicted with a reasonable degree of accuracy, if so how?

2. Can the outcome of a meet be predicted with a reasonable degree of accuracy, given project swimmer times.

3. Is there a better way to rank recruits in order to pair them with schools where they will have the greatest degree of success.

Answering these questions will require meticulous aggregation of the relevant data, and detailed and precise exploration of the data before any large scale analysis is done. Some important questions that lie at the core of the above 3 are:

* Whether past times can be used to predict future times

* Whether times within a team can help predict teammates times

* Whether a network mapping of club and collegiate swim teams with weighted edges for successful collegiate swimmers leads to any relevant findings.

Power Points are a metric that was designed by USA Swimming to quantify how rare a time is for a certain gender, age group, and event. We need to read in the data and transform some of the data to make it easier to work with.  Currently most of the values are strings, and we need to convert the Time2 column to seconds, and the Date column to a Datetime object.

```python
# Read data into DB.
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as dates
import time
import datetime

# convert listed time to absolute seconds.
def getSec(time):
	if ':' in time:
		l = time.split(':')
		return int(l[0]) * 60 + float(l[1])
	else:
		return float(time)

# loads in the results to pandas and add column names
df = pd.read_csv("results_clean.txt", header=None, delimiter = '\t', error_bad_lines = False)
df.columns = ('Name','Team','Year','School','Event','Course','Age','Time1','Time2','Power Points','Standard','Meet','LSC','Club','Date')

# cleans time into seconds
df['Time2'] = df['Time2'].apply(lambda x: getSec(x))
df['Date'] = df['Date'].apply(lambda x: datetime.datetime.strptime(x, "%m/%d/%Y"))

print df.head()
```

```python
"""
Function taken from the CS109 Homeworks
Function
--------
remove_border
removes the border of graphs

Inputs
------
None
    
Returns
-------
Nothing
"""

%matplotlib inline
import matplotlib.pyplot as plt

def remove_border(axes=None, top=False, right=False, left=True, bottom=True):
    """
    Minimize chartjunk by stripping out unnecessary plot borders and axis ticks
    
    The top/right/left/bottom keywords toggle whether the corresponding plot border is drawn
    """
    ax = axes or plt.gca()
    ax.spines['top'].set_visible(top)
    ax.spines['right'].set_visible(right)
    ax.spines['left'].set_visible(left)
    ax.spines['bottom'].set_visible(bottom)
    
    #turn off all ticks
    #ax.yaxis.set_ticks_position('none')
    #ax.xaxis.set_ticks_position('none')
    
    #now re-enable visibles
    if top:
        ax.xaxis.tick_top()
    if bottom:
        ax.xaxis.tick_bottom()
    if left:
        ax.yaxis.tick_left()
    if right:
        ax.yaxis.tick_right()
```

```python
"""
Function
--------
plot_times
plots the time progressions of all swimmers given.

Inputs
------
swim_sub: Panda dataframe subset of the swimmers we would like to compare (just subselect names from the dataframe)
race: Event (i.e. "50 FR", or "200 BR")
course: "SCY" or "LCM"
    
Returns
-------
Nothing
"""

def plot_times(swim_sub, course, event):
    # generate a list of unique names from the dataset.
    swimmers = list(set(swim_sub.Name))
    xmin, xmax, ymin, ymax = [], [], [], []
    # plot each swimmer's times and save min values for axis plotting.
    for swimmer in swimmers:
        subdf = swim_sub[(swim_sub["Name"] == swimmer) & (swim_sub["Course"] == course) & (swim_sub["Event"] == event)]
        subdf = subdf.sort('Date')
        plt.plot(dates.date2num(subdf.Date), subdf.Time2, label=swimmer)
        xmin.append(min(subdf.Date))
        xmax.append(max(subdf.Date))
        ymin.append(min(subdf.Time2))
        ymax.append(max(subdf.Time2))
    # axis formatting
    plt.axis([min(xmin)-datetime.timedelta(days=30), max(xmax)+datetime.timedelta(days=30), min(ymin)-((max(ymax)-min(ymin))*.05), max(ymax)+((max(ymax)-min(ymin))*.05)])
    plt.xlabel("Date")
    plt.ylabel("Time (seconds)")
    plt.title("Time progressions for: " + event + " - " + course)
    remove_border()
    # legend
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    plt.show()
```

#### Moving Average & Curve Fitting

```python
"""
Function
--------
sma
simple moving average of past times k times

Inputs
------
swim_subi: Subset of data for an individual swimmer
race: Event (i.e. "50 FR", or "200 BR")
course: "SCY" or "LCM"
k: the number of past races to include in the average
    
Returns
-------
A modified copy of the original individual subset so it can be plotted.
"""

def sma(swim_subi, course, event, k):
    swim_subi = swim_subi[(swim_subi["Course"] == course) & (swim_subi["Event"] == event)]
    swim_subi = swim_subi.sort('Date')
    # holder for average times
    sma_times = []
    # exit if not enough data
    if (len(swim_subi.Time2) < k):
        return None
    else:
        swims = list(swim_subi.Time2)
        # compute the average of the past k swims
        for i in range(len(swims) - k):
            sma_times.append(sum(swims[i:i+k])/float(k))
        # copy dataframe and insert new times
        sma_cop = swim_subi[k:].copy(deep=True)
        sma_cop.Time2 = sma_times
        # create a new name to compare to other times
        sma_cop.Name = sma_cop.Name + "_sma" + "_" + str(k)
        return sma_cop
```

```python
# Let's try our moving average and compare it with the
# raw data from some of the swimmers we examined above.
cls = df[(df["Name"] == "Christopher Satterthwaite")]
cls_sma = sma(cls, "SCY", "50 FR", 3)
plot_times(cls.append(cls_sma), "SCY", "50 FR")
cls_sma = sma(cls, "SCY", "50 FR", 6)
plot_times(cls.append(cls_sma), "SCY", "50 FR")
```

As we can see above, the simple moving average does help to eliminate some of the noise, but it reduces the amount of data we have to work with, and it doesn't give us a clear picture of what the future holds. By definition, the average can never be faster than the fastest time of a swimmer, so we get no predictive benefit from the moving average, and it is not immediately apparent how we could extrapolate the rate of descent to predict future times.

Part of the problem is likely due to the fact that in season times are so much slower than championship times, and championship times are what matter most.  To get around this problem, we will see if we can tease apart championship times and in season times and analyze them seperately.  If we notice consistent rates of descent, then perhaps we have a reliable model for predicting future times.

```python
"""
Function
--------
championship_sep
separates in season and championship times and returns a modified dataframe

Inputs
------
swim_subi: Subset of data for an individual swimmer
race: Event (i.e. "50 FR", or "200 BR")
course: "SCY" or "LCM"
    
Returns
-------
A modified copy of the original individual subset so it can be plotted.
"""

def championship_sep(swim_subi, course, event):
    swim_subi = swim_subi[(swim_subi["Course"] == course) & (swim_subi["Event"] == event)]
    swim_subi = swim_subi.sort('Date')
    # holder for times
    champ_times = []
    avg_times = []
    # exit if not enough data
    if (len(swim_subi.Time2) < 2):
        return None
    else:
        swims = list(swim_subi.Time2)
        swim_years = list(set(swim_subi["Date"].apply(lambda x: x.year)))
        for yr in swim_years:
            sub_yr = swim_subi[swim_subi["Date"].apply(lambda x: x.year) == yr]
            if (len(sub_yr.Time2) == 1):
                champ_times.append(list(sub_yr.Time2)[0])
                avg_times.append(list(sub_yr.Time2)[0])
            else:
                champ_times.append(min(sub_yr.Time2))
                avg_times.append((sum(sub_yr.Time2) - min(sub_yr.Time2))/(len(sub_yr.Time2)-1.0))
        # create a new avg dataframe to contain times
        new_times_a = swim_subi.copy(deep=True)
        new_times_a = new_times_a[:len(swim_years)]
        new_times_a["Name"] = swim_subi.Name + "_avg"
        new_times_a["Time2"] = avg_times
        new_times_a["Date"] = swim_years
        new_times_a["Age"] = new_times_a["Time1"] = new_times_a["Power Points"] = new_times_a["Standard"] = new_times_a["Meet"] = new_times_a["LSC"] = new_times_a["Club"] = ""
        # create a new champ dataframe
        new_times_c = new_times_a.copy(deep=True)
        new_times_c["Name"] = swim_subi.Name + "_champ"
        new_times_c["Time2"] = champ_times
        new_times = new_times_a.append(new_times_c)
        # convert times back to datetime format
        new_times["Date"] = new_times["Date"].apply(lambda x: datetime.datetime.strptime(str(x), "%Y"))
        return new_times
```

```python
"""
Function
--------
swimpred
predict Ivy 2014 championship times (simple)

Inputs
------
pred_fun: prediction function to fit, something of the form a+b*x+c*x*x
indiv_sub: individual subset of the dataframe (raw times only)
race: Event (i.e. "50 FR", or "200 BR")
course: "SCY" or "LCM"
    
Returns
-------
A predicted time for a given event.  No variation in the data.
"""

# import module to optimize curve
import scipy.optimize as optimization

# return forecast time for swimmer.
def swimpred(pred_fun, indiv_sub, course, event):
    # IVY DATE
    ivy14 = dates.date2num(datetime.datetime.strptime('3/1/2014', '%m/%d/%Y'))
    # subset & sort
    indiv_sub = indiv_sub[(indiv_sub["Course"] == course) & (indiv_sub["Event"] == event)]
    indiv_sub = indiv_sub.sort('Date')
    # do not predict swims with less than 6 times or if most recent swim is over a year old. 
    if ((len(indiv_sub.Time2) < 6) or (max(indiv_sub.Date) < datetime.datetime.strptime('3/1/2012', '%m/%d/%Y'))):
        return None
    # compile data for fitting function
    xdata = np.array(dates.date2num(indiv_sub.Date))
    ydata = np.array(indiv_sub.Time2)
    # fit function to all in season data
    popt, pcov = optimization.curve_fit(func, xdata, ydata)
    # predict aggregate time
    pred_time_all = pred_fun(ivy14, *popt)
    # championship data for swimmer
    champdat = championship_sep(indiv_sub, course, event)
    champdat = champdat[champdat['Name'] == list(set(indiv_sub.Name))[0] + '_champ']
    champdat = champdat.sort('Date')
    champdates = np.array(dates.date2num(champdat.Date))
    if (len(champdat.Time2) >= 3):
        # fit function to championship data
        popt, pcov = optimization.curve_fit(pred_fun, champdates, champdat.Time2)
        # predict championship time
        pred_time_champ = pred_fun(ivy14, *popt)
        # return weighted combination of both
        return pred_time_champ
    else:
        return pred_time_all
```

The function below takes an individuals event subset of the dataframe, and calculates the average amount dropped over the past 8 best times, and also the standard deviation among drops.  The output of this function will be used to randomly sample a number of seconds to drop, with a normal distribution centered on the mean of dropped time and with the respective variance.

```python
"""
Function
--------
pred_fun_error
predict an event's time based on past drops and drop variation

Inputs
------
df_indiv: individual subset of the dataframe (raw times only) for a certain event
    
Returns
-------
fastest time in event, mean of past 7 drops, standard deviation of past 7 drops
"""

def pred_fun_error(df_indiv):
    df_indiv = df_indiv.sort('Date')
    # drop if not enough data or not recent enough
    if ((len(df_indiv.Time2) < 8) or (max(df_indiv.Date) < datetime.datetime.strptime('3/1/2012', '%m/%d/%Y'))):
        return None
    df_indiv = df_indiv.sort('Time2')
    fast_times = list(df_indiv[:8].Time2)
    ctr = []
    # calculate drops in past swims
    for i in range(len(fast_times)-1):
        ctr.append(fast_times[i+1]-fast_times[i])
    fast_mean = np.mean(ctr)
    fast_sd = np.std(ctr)
    fast = min(fast_times)
    # return calculated metrics
    return fast, fast_mean, fast_sd
```

It takes a while to generate all the means and standard deviations, and since the calculations won't change, we want to save this output so we can repeatedly use the information to sample from a distribution if we want to run a meet multiple times (each time we would have slightly different predicted times for each swimmer).  This allows us to examine variability in the outcome of a projected swim meet, and we can see how robust a designated lineup is to changes in times of that team and different teams.

The function below calulcates the parameters for the drop distributions for the entire dataset.

```python
from scipy import stats

"""
Function
--------
pred_var
predict drop parameters

Inputs
------
pred_fun_eror: function to return parameters for normal drop dist
dataset: full dataset to compute drop params for
    
Returns
-------
dataset with fastest time, mean, and var of drops.
"""

def pred_var(pred_fun_error, dataset):
    # blank PD row template
    blank_row = pd.DataFrame({'Name' : "", 'Team' : "", 'Year' : "", 'School' : "", 'Event' : "", 'Course' : "", 'Age' : "", 'Time1' : "", 'Time2' : "", 'Power Points' : "", 'Standard' : "", 'Meet' : "", 'LSC' : "", 'Club' : "", 'Date' : ""}, index=['TODO'])
    # valid Ivy events
    events = ["50 FR", "100 FR", "200 FR", "500 FR", "1000 FR", "1650 FR", "100 BK", "200 BK", "100 BR", "200 BR", "100 FL", "200 FL", "200 IM", "400 IM"]
    # initialize empty DataFrame
    df_pred_big = blank_row.copy()
    # iterate through each person
    for name in list(set(dataset.Name)):
        df_pred_person = blank_row.copy()
        df_sub = dataset[dataset["Name"] == name]
        # iterate through each event
        for event in events:
            new_sub = df_sub[(df_sub['Event'] == event) & (df_sub['Course'] == "SCY")]
            # get parameters for normal distributions
            pred_out = pred_fun_error(new_sub)
            if (pred_out is not None):
                best, pred_mean, pred_sd = pred_out
                new_row = blank_row.copy()
                new_row["Event"] = event
                new_row["Time2"] = best
                new_row["Mean"] = pred_mean
                new_row["SD"] = pred_sd
                # add time to existing dataframe
                df_pred_person = df_pred_person.append(new_row)
        # add swimmer dataframe to aggregate frame
        df_pred_person["Name"] = name
        df_pred_person['School'] = list(df_sub['School'])[0]
        df_pred_person['Year'] = list(df_sub['Year'])[0]
        df_pred_person = df_pred_person[df_pred_person['Event'] != ""]
        df_pred_big = df_pred_big.append(df_pred_person)
    # return prediction dataframe
    df_pred_big = df_pred_big[df_pred_big['Event'] != ""]
    df_pred_big.index = range(len(df_pred_big.Time2))
    return df_pred_big
```

The function below takes the original dataset, and the parameters of the drops for each event, and it returns a dataset of predicted times which are the fastest time plus an amount sampled from the dropping distribution.  This is where we introduce the variability that allows us to examine variability in the meet projection model.

```python
"""
Function
--------
pred_werror
predict times for entire dataset based on sampled drops

Inputs
------
dataset: entire dataset
error_df: dataset of means and variances, generated from pred_var()
    
Returns
-------
dataset of predicted times (same as our curve fitting model)
"""

def pred_werror(dataset, error_df):
    # blank PD row template
    blank_row = pd.DataFrame({'Name' : "", 'Team' : "", 'Year' : "", 'School' : "", 'Event' : "", 'Course' : "", 'Age' : "", 'Time1' : "", 'Time2' : "", 'Power Points' : "", 'Standard' : "", 'Meet' : "", 'LSC' : "", 'Club' : "", 'Date' : ""}, index=['TODO'])
    # valid Ivy events
    events = ["50 FR", "100 FR", "200 FR", "500 FR", "1000 FR", "1650 FR", "100 BK", "200 BK", "100 BR", "200 BR", "100 FL", "200 FL", "200 IM", "400 IM"]
    # initialize empty DataFrame
    df_pred_big = blank_row.copy()
    # iterate through each person
    for name in list(set(dataset.Name)):
        df_pred_person = blank_row.copy()
        df_sub = dataset[dataset["Name"] == name]
        df_error_sub = error_df[error_df['Name'] == name]
        # iterate through each event
        for event in events:
            new_sub = df_error_sub[(df_error_sub['Event'] == event)]
            if (len(list(new_sub.Time2)) > 0):
                # calculate predicted time with sampled drop
                pred_time = stats.norm.rvs(-(list(new_sub.Mean)[0]), list(new_sub.SD)[0], 1)[0] + list(new_sub.Time2)[0]
                new_row = blank_row.copy()
                new_row["Event"] = event
                new_row["Time2"] = pred_time
                # add time to existing dataframe
                df_pred_person = df_pred_person.append(new_row)
            else:
                continue
        # add swimmer dataframe to aggregate frame
        df_pred_person["Name"] = name
        df_pred_person['School'] = list(df_sub['School'])[0]
        df_pred_person['Year'] = list(df_sub['Year'])[0]
        df_pred_person = df_pred_person[df_pred_person['Event'] != ""]
        df_pred_big = df_pred_big.append(df_pred_person)
    # return prediction dataframe
    df_pred_big = df_pred_big[df_pred_big['Event'] != ""]
    df_pred_big.index = range(len(df_pred_big.Time2))
    return df_pred_big
```

To predict Ivy's, we will need to determine what event each swimmer should swim as we have their times from every event they have swam. The original thought was to just do a linear ranking of times of all the people from that year, and the swimmer should swim whichever they were the highest in. However, that made it difficult to compare large differences in times if there was no one else in between them (22.0 and 22.5). Instead, we pulled all the top times from that each person has swam in each event. Then, given an event and time, we can see how good that swim would be relative to the history of the Ivy League. That way we can better which swims are more impressive, and thus the swimmer should choose to swim that event as opposed to another they may be able to swim. After pulling the best times from everyone in every event, the function below calculates the percentiles.

```python
# LOAD TOP TIMES IN EACH EVENT
# get unique names and events
names = list(set(df.Name))
events = list(set(df.Event))
# store all of the top times
top_times = {}

# for each event for each swimmer
for name in names:
    for event in events:
        # if they have a best time, add to list of top times
        if not df[(df.Name == name) & (df.Event == event) & (df.Course == "SCY")].empty:
            if top_times.has_key(event):
                top_times[event].append(min(df[(df.Name == name) & (df.Event == event) & (df.Course == "SCY")].Time2))
            else:
                top_times[event] = [min(df[(df.Name == name) & (df.Event == event) & (df.Course == "SCY")].Time2)]



"""
Function
--------
rank_swim

Inputs
------
time: a time you are looking to see how good it is
race: Event (i.e. "50 FR", or "200 BR")
    
Returns
-------
The percentile ranking of that swim in the ivy league (a number from 0 to 100)
"""
from scipy import stats
import scipy as sp
def rank_swim(time, race):
    if top_times.has_key(race):
        return 100 - stats.percentileofscore(top_times[race], time)
    else:
        return 0
```

Now that we have a way to rank swims and determine which one may be better than the other (as you can't just compare the times of a 1650 yard race and a 200 yard race), we can use that information to pick an Ivy team. The function below takes a Pandas data frame generated from above that lists all the possible swimmers, events, and their expected times and then picks 17 swimmers and their events for each team. It simply ranks every swimmer's events, then picks the top three. Then, to choose the 17 for the team, it picks the top 17 swimmers who have the highest total ranking score of their three events.

```python
"""
Function
--------
pick_ivy_teams

Inputs
------
data: a Pandas dataframe with swimmers, team, events, and expected times.
    
Returns
-------
A pandas dataframe with only 17 swimmers per team and three events per swimmer
"""

def pick_ivy_teams(data):
    # create results dataframe
    results = pd.DataFrame(columns=["Swimmer", "School", "Event", "Time"])
    schools = sorted(list(set(data.School)))
    
    # pick each team
    for school in schools:
        if school == "":
            continue
        ivyteam = []
        # for each swimmer, calculate ranking of events and total ranking
        for swimmer in list(set(data[data.School == school].Name)):
            swimmer_events = []
            for event in list(set(data[(data.Name == swimmer) & (data.School == school)].Event)):
                time = data[(data.Name == swimmer) & (data.School == school) & (data.Event == event)].Time2.values[0]
                swimmer_events.append((event, time, rank_swim(time, event)))
            # only three swims
            best_swims = sorted(swimmer_events, key=lambda tup: -tup[2])[:3]
            # overall "power ranking" of those three swims
            overall_power_ranking = sum(x[2] for x in best_swims)
            ivyteam.append([swimmer, overall_power_ranking, [(x[0], x[1]) for x in best_swims]])
        # only top 17 swimmers
        swimmers = sorted(ivyteam, key=lambda tup: -tup[1])[:17]
        
        # add each swimmer to the results dataframe
        for swimmer in swimmers:
            for race in swimmer[2]:
                results = results.append({"Swimmer": swimmer[0], "School": school, "Event": race[0], "Time": race[1]}, ignore_index=True)
                
    return results.sort(["Event", "Time"])
```

But that may not be the best way to choose which events a person should swim. One team may choose to move someone to another event even if they won't place as high because it could still be a net positive. For example, there may be 5 Harvard swimmers and 2 Princeton swimmers in the mile and 4 Princeton swimmers in the 200 fly and only 1 Harvard swimmer in the 200 fly. It is possible that the top Harvard swimmer in the mile would have won that event but if he were to swim the 200 fly instead he could beat enough Princeton swimmers there to take points away from them and there would still be Harvard swimmers in the mile to get those points. Thus, it would be worth it to swim an event you may not place as high in if it will move other teams down and your team still has swimmers in that event to keep the points. But if Princeton knows Harvard does that, they may change their line up to counter that. We now need to look at [game theory](http://en.wikipedia.org/wiki/Game_theory) to figure out which lineups each team would (or should) choose, since we are assuming the coach for each team is as good as all the others.

The process of each team choosing the lineup becomes an 8 person [zero-sum](http://en.wikipedia.org/wiki/Zero%E2%80%93sum_game) (constant-sum) [simultaneuos](http://en.wikipedia.org/wiki/Simultaneous_game) [non-cooperative](http://en.wikipedia.org/wiki/Non-cooperative_game) game. These types of games are theoretically solvable if there exists a [Nash Equilibrium](http://en.wikipedia.org/wiki/Nash_equilibrium). A Nash equilibrium is a solution concept to a game such that each player has chosen a strategy, and one person cannot benefit by changing their strategy while no one else changes their's. In our case, a strategy for a team is a given line up, and you measure the success of that lineup by scoring out the meet.

The function below calculates all Nash Equilibrium points given a list of possible lineups for each team using a strategy called Best Response analysis. After having a full payoff matrix (an 8 dimensional matrix with the scores of a meet from every lineup), you find each team's best response for each possible set of the other seven team's line ups. Then, after finding every team's best response to every possible set of other team's lineups, if there are any points in the matrix that are the best response for all 8 teams, then that is a Nash equilibrium. The code below finds any and all of them.


```python
"""
Function
--------
nash_eq

Inputs
------
brown, col, corn, harv, penn, prin, yale: list of lineups 
    
Returns
-------
the lineups for the nash equilibrium(s)
"""

def nash_eq(brown, col, corn, dart, harv, penn, prin, yale):
    # find number of "strategies" (line ups) for each team
    brown_n = len(brown)
    col_n = len(col)
    corn_n = len(corn)
    dart_n = len(dart)
    harv_n = len(harv)
    penn_n = len(penn)
    prin_n = len(prin)
    yale_n = len(yale)
    
    # initialize 8 dimensional score array with zeroes
    scores = np.zeros((brown_n, col_n, corn_n, dart_n, harv_n, penn_n, prin_n, yale_n), 
                      dtype='u2, u2, u2, u2, u2, u2, u2, u2')
    
    # fill scores array with scores from those lineups
    for b in range(0,brown_n):
        for cl in range(0,col_n):
            for cn in range(0,corn_n):
                for d in range(0,dart_n):
                    for h in range(0,harv_n):
                        for pe in range(0,penn_n):
                            for pr in range(0,prin_n):
                                for y in range(0,yale_n):
                                    meet = pd.concat([brown[b],col[cl],corn[cn],dart[d],harv[h],penn[pe],prin[pr],yale[y]])
                                    points = tuple([x[1] for x in score_ivys(meet)])
                                    scores[b][cl][cn][d][h][pe][pr][y] = points
    
    ### begin best response analysis
    # brown
    brown_nash = []
    for cl in range(0,col_n):
        for cn in range(0,corn_n):
            for d in range(0,dart_n):
                for h in range(0,harv_n):
                    for pe in range(0,penn_n):
                        for pr in range(0,prin_n):
                            for y in range(0,yale_n):
                                # given other team's strategies, find this one's best response
                                max_score = 0
                                nash = []
                                for b in range(0,brown_n):
                                    score = scores[b][cl][cn][d][h][pe][pr][y]
                                    if score > max_score:
                                        max_score = score
                                        nash = [(b,cl,cn,d,h,pe,pr,y)]
                                    elif score == max_score:
                                        nash.append((b,cl,cn,d,h,pe,pr,y))
                                # keep track of each best response
                                brown_nash = brown_nash + nash
                                
    # columbia
    col_nash = []
    for b in range(0,brown_n):
        for cn in range(0,corn_n):
            for d in range(0,dart_n):
                for h in range(0,harv_n):
                    for pe in range(0,penn_n):
                        for pr in range(0,prin_n):
                            for y in range(0,yale_n):
                                # given other team's strategies, find this one's best response
                                max_score = 0
                                nash = []
                                for cl in range(0,col_n):
                                    score = scores[b][cl][cn][d][h][pe][pr][y]
                                    if score > max_score:
                                        max_score = score
                                        nash = [(b,cl,cn,d,h,pe,pr,y)]
                                    elif score == max_score:
                                        nash.append((b,cl,cn,d,h,pe,pr,y))
                                # keep track of each best response
                                col_nash = col_nash + nash
    
    # cornell
    corn_nash = []
    for b in range(0,brown_n):
        for cl in range(0,col_n):
            for d in range(0,dart_n):
                for h in range(0,harv_n):
                    for pe in range(0,penn_n):
                        for pr in range(0,prin_n):
                            for y in range(0,yale_n):
                                # given other team's strategies, find this one's best response
                                max_score = 0
                                nash = []
                                for cn in range(0,corn_n):
                                    score = scores[b][cl][cn][d][h][pe][pr][y]
                                    if score > max_score:
                                        max_score = score
                                        nash = [(b,cl,cn,d,h,pe,pr,y)]
                                    elif score == max_score:
                                        nash.append((b,cl,cn,d,h,pe,pr,y))
                                # keep track of each best response
                                corn_nash = corn_nash + nash
                                
    # dartmouth 
    dart_nash = []
    for b in range(0,brown_n):
        for cl in range(0,col_n):
            for cn in range(0,corn_n):
                for h in range(0,harv_n):
                    for pe in range(0,penn_n):
                        for pr in range(0,prin_n):
                            for y in range(0,yale_n):
                                # given other team's strategies, find this one's best response
                                max_score = 0
                                nash = []    
                                for d in range(0,dart_n):
                                    score = scores[b][cl][cn][d][h][pe][pr][y]
                                    if score > max_score:
                                        max_score = score
                                        nash = [(b,cl,cn,d,h,pe,pr,y)]
                                    elif score == max_score:
                                        nash.append((b,cl,cn,d,h,pe,pr,y))
                                # keep track of each best response
                                dart_nash = dart_nash + nash
                                    
    # harvard
    harv_nash = []
    for b in range(0,brown_n):
        for cl in range(0,col_n):
            for cn in range(0,corn_n):
                for d in range(0,dart_n):
                    for pe in range(0,penn_n):
                        for pr in range(0,prin_n):
                            for y in range(0,yale_n):
                                # given other team's strategies, find this one's best response
                                max_score = 0
                                nash = [] 
                                for h in range(0,harv_n):
                                    score = scores[b][cl][cn][d][h][pe][pr][y]
                                    if score > max_score:
                                        max_score = score
                                        nash = [(b,cl,cn,d,h,pe,pr,y)]
                                    elif score == max_score:
                                        nash.append((b,cl,cn,d,h,pe,pr,y))
                                # keep track of each best response
                                harv_nash = harv_nash + nash
    
    # penn
    penn_nash = []
    for b in range(0,brown_n):
        for cl in range(0,col_n):
            for cn in range(0,corn_n):
                for d in range(0,dart_n):
                    for h in range(0,harv_n):
                        for pr in range(0,prin_n):
                            for y in range(0,yale_n):
                                # given other team's strategies, find this one's best response
                                max_score = 0
                                nash = [] 
                                for pe in range(0,penn_n):
                                    score = scores[b][cl][cn][d][h][pe][pr][y]
                                    if score > max_score:
                                        max_score = score
                                        nash = [(b,cl,cn,d,h,pe,pr,y)]
                                    elif score == max_score:
                                        nash.append((b,cl,cn,d,h,pe,pr,y))
                                # keep track of each best response
                                penn_nash = penn_nash + nash
                                
    # princeton
    prin_nash = []
    for b in range(0,brown_n):
        for cl in range(0,col_n):
            for cn in range(0,corn_n):
                for d in range(0,dart_n):
                    for h in range(0,harv_n):
                        for pe in range(0,penn_n):
                            for y in range(0,yale_n):
                                # given other team's strategies, find this one's best response
                                max_score = 0
                                nash = []
                                for pr in range(0,prin_n):
                                    score = scores[b][cl][cn][d][h][pe][pr][y]
                                    if score > max_score:
                                        max_score = score
                                        nash = [(b,cl,cn,d,h,pe,pr,y)]
                                    elif score == max_score:
                                        nash.append((b,cl,cn,d,h,pe,pr,y))
                                # keep track of each best response
                                prin_nash = prin_nash + nash
    
    # yale
    yale_nash = []
    for b in range(0,brown_n):
        for cl in range(0,col_n):
            for cn in range(0,corn_n):
                for d in range(0,dart_n):
                    for h in range(0,harv_n):
                        for pe in range(0,penn_n):
                            for pr in range(0,prin_n):
                                # given other team's strategies, find this one's best response
                                max_score = 0
                                nash = []
                                for y in range(0,yale_n):
                                    score = scores[b][cl][cn][d][h][pe][pr][y]
                                    if score > max_score:
                                        max_score = score
                                        nash = [(b,cl,cn,d,h,pe,pr,y)]
                                    elif score == max_score:
                                        nash.append((b,cl,cn,d,h,pe,pr,y))
                                # keep track of each best response
                                yale_nash = yale_nash + nash
                                
    # find any and all nash equilibriums                  
    return list(set(corn_nash) & set(col_nash) & set(dart_nash) & set(harv_nash) & set(penn_nash) & set(prin_nash) & set(yale_nash))
```
